```python
import pandas as pd
from presidio_analyzer import AnalyzerEngine

import spacy
import spacy.cli

from flair.data import Sentence
from flair.models import SequenceTagger

from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
```

```python
dessi_test = pd.read_csv("/home/aagisha/Documents/PhD Albert/CodingPhDStart1/Bachelor Projects HSAA/PIIDETECTION/Personal-Detection/datasets/dessi/DeSSI_v2/test.csv")
```

```python
def convert_to_text(df, row):
    s = "This data point has the following values: "
    for c in df.columns:
        s = s + str(c) + " " + str(df.loc[row,c]) + ",\n"
    return s
```

```python
dessi_string = convert_to_text(dessi_test, 0)
```

```python
dessi_string
```

# Microsoft Presidio

```python
analyzer = AnalyzerEngine()

results = analyzer.analyze(text=dessi_string, entities=[], language='en')

for result in results:
    print(f"Entity: {result.entity_type}, Start: {result.start}, End: {result.end}, Score: {result.score}, Text: {dessi_string[result.start:result.end]}")
```

# spacy

```python
# Das Modell "en_core_web_sm" herunterladen
spacy.cli.download("en_core_web_sm")

# Laden des vortrainierten Modells
nlp = spacy.load("en_core_web_sm")



doc = nlp(dessi_string)

for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")
```

# Flair

```python
tagger = SequenceTagger.load("ner")


sentence = Sentence(dessi_string)
tagger.predict(sentence)


for entity in sentence.get_spans('ner'):
    print(f"Entity: {entity.text}, Label: {entity.get_label('ner').value}")
```

## HuggingFace Transformers

```python
model_name = "xlm-roberta-large-finetuned-conll03-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)


nlp = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")


entities = nlp(dessi_string)


for entity in entities:
    print(f"Entity: {entity['word']}, Label: {entity['entity_group']}, Score: {entity['score']}")
```
